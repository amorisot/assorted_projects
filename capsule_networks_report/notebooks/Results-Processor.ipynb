```python
import csv
import 
import os
```

```python
LOCATION = '../experiment_analysis/CW3_results/provided_cifar10_results/'
```

```python
VAL_ACC_KEY = 'val_c_accuracy'
```

```python
results = {}
for filename in os.listdir(LOCATION):
    with open(os.path.join(LOCATION, filename)) as f:
        reader = csv.DictReader(f)
        results[filename] = list(reader)
```

## Look at Early Stopping options

```python
EARLY_STOP_AFTER = 10
```

```python
for filename, data in results.items():
    print()
    print(filename)
    remaining_epochs = EARLY_STOP_AFTER
    best_acc = 0
    for epoch_data in data[:-1]:  # exclude the last line
        curr_val_acc = float(epoch_data[VAL_ACC_KEY])
        did_improve = best_acc < curr_val_acc
        if did_improve:
            best_acc = curr_val_acc
            remaining_epochs = EARLY_STOP_AFTER
        else:
            remaining_epochs -= 1
            if remaining_epochs <= 0:
                pass # can break here if you want to early stop
        print('\t'.join(map(str, (epoch_data['epoch'], curr_val_acc, '(improved)' if did_improve else ''))))
    
```
