# Plot the data :D

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import os
```

```python
# change this if the notebook is moved the /notebooks folder

LOCATION = 'shuffled_capsnet/epoch_stats.csv'
```

```python
# this is for training and validation accuracies


data = np.genfromtxt(LOCATION, delimiter=',', skip_header = 1, usecols = (0, 4, 6))


epochs = data[:,0]
train_acc = data[:,1]*100
val_acc = data[:,2]*100


#plotting part
train, = plt.plot(epochs, train_acc, label='Training Accuracy', color='blue')
val, = plt.plot(epochs, val_acc, label='Validation Accuracy', color='red')
plt.legend(handles=[train, val])
plt.xlabel('Number of Epochs')
plt.ylabel('Accuracy (in %)')
plt.show()
#save with name and in location of choice using this command, preferably as an .svg
#plt.savefig('cifar10.svg')
```
